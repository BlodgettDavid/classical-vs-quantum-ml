# 🧠 Classical MLP Classifier on Simple Dataset

-This notebook trains a classical neural network (MLP) on the `make_moons` dataset and evaluates its performance. It serves as a baseline for comparison with quantum models in later notebooks.

```python
# imports

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
# Seed and Data Generation

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

# Generate the make_moons dataset
X, y = make_moons(n_samples=200, noise=0.1, random_state=RANDOM_SEED)
```

```python
# Scaling

# Scale features to [0, π] for compatibility with quantum models later
scaler = MinMaxScaler(feature_range=(0, np.pi))
X_scaled = scaler.fit_transform(X)
```

```python
# Train/Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=RANDOM_SEED)
```

```python
# Classical MLP Model

mlp = MLPClassifier(
    hidden_layer_sizes=(10, 10),
    max_iter=500,
    random_state=RANDOM_SEED
)
mlp.fit(X_train, y_train)
```

```python
# Evaluation

y_pred_mlp = mlp.predict(X_test)
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)

print("-" * 50)
print(f"✅ Classical MLP Accuracy: {accuracy_mlp*100:.2f}%")
print("-" * 50)
```

```python
# Visualization

plt.figure(figsize=(6, 4))
plt.scatter(X_scaled[y == 0, 0], X_scaled[y == 0, 1], label='Class 0', marker='o')
plt.scatter(X_scaled[y == 1, 0], X_scaled[y == 1, 1], label='Class 1', marker='x')
plt.title("Make Moons Dataset (Scaled)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
